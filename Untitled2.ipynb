## cleaning the data

```python
# fill the empty cells in ('gross','runtime','rating','votes')
df['gross']=df['gross'].fillna(df['gross'].median())
df['runtime']=df['runtime'].fillna(df['runtime'].median())
df['rating']=df['rating'].fillna(np.random.choice(['R', 'PG', 'G', 'Not Rated', 'NC-17', 'TV-PG',
       'PG-13', 'Unrated', 'X', 'TV-MA', 'TV-14']))
df['votes']=df['votes'].fillna(df['votes'].mean())
```

```python
new_budget=df.dropna(subset=['budget'])['budget']
df['budget']=df['budget'].apply(lambda x: np.random.choice(new_budget)
                               if pd.isnull(x) else x)
```

```python
df=df.dropna()
```

```python
# checking duplicated rows
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
# change the dtypes (votes,budget,gross,runtime) columns
df['votes']=df['votes'].astype('int64')
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
df['runtime']=df['runtime'].astype('int64')
```

```python
df.info()
```

```python
# order or values a little bit to organize it
df=df.sort_values('gross',ascending=False).reset_index(drop=True)
df
```
